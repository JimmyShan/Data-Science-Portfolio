# Get Amenity list
---

This script collects list of amenities from all apartment complex. The collected amenty list is used in 'data_cleaning' to convert amenities from a string to dummy variables.

Import basic library

```python
import pandas as pd
import numpy as np
```

Read data as csv and store it is a data frame

```python
df = pd.read_csv("data/ny_rental_data.r3.csv", index_col = 0)
```

Define set to hold the list of amenties

```python
amenties_set = set()
```

```python
def create_amenty_list(amenty):
    # remove un-wanted characters
    amenty = amenty.replace('[','')
    amenty = amenty.replace(']','')
    amenty = amenty.replace("'","")

    # create list of amenties from the string
    amenty_list = amenty.split(",")

    for a in amenty_list:
        amenties_set.add(a.lower())

    return amenty_list
```

apply the function to amenty column to get list of amenties

```python
df["amenty"].apply(create_amenty_list)
```

print amenty and its count

```python
from collections import Counter
for amenty in Counter(amenties_set).most_common():
    print(amenty)
    print(amenty[0])
```
